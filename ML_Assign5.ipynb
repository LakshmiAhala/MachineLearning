```python
#@title Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

# Setting display options
pd.set_option('display.max_columns', None)
%matplotlib inline

print("‚úì Libraries imported successfully!")
```

#Part1: SETUP AND EXAMPLE

```python
#@title 1.1: Loading Iris Dataset

# Step 1: Loading the dataset
iris = datasets.load_iris()
print("Dataype of iris: ", type(iris))

# Step 2: Understanding the dataset structure
print("üì¶ Dataset Components:")
print(f"   - data: Features (measurements)")
print(f"   - target: Labels (species)")
print(f"   - feature_names: Column names for features")
print(f"   - target_names: Names of species")
print(f"   - DESCR: Dataset description\n")

# Step 3: Creating DataFrame from features
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
print("dataype of df_iris: ", type(df_iris))

# Step 4: Adding the target column
df_iris['species'] = iris.target

# Step 5: Replacing numeric df_iris with actual names
df_iris['species_name'] = df_iris['species'].map({
    0: iris.target_names[0],
    1: iris.target_names[1],
    2: iris.target_names[2]
})

print("‚úÖ iris Dataset loaded successfully!")
print(f"\nDataFrame shape: {df_iris.shape}")
print(f"Columns: {list(df_iris.columns)}")

# Displaying the first few rows
print("üîç First 5 rows of Iris dataset:\n")
display(df_iris.head())

print("\nüìä Basic Information:")
df_iris.info()
```

```python
#@title 1.2: Read the Dataset Description

descr = iris.DESCR
print(descr)
```

#Part3: BREAST CANCER DATASET EXPLORATION

```python
#@title 3.1: Loading Breast Cancer Dataset

# Step 1: Loading the breast cancer dataset
bc = datasets.load_breast_cancer()

# Step 2: Creating a DataFrame with the features
df_bc = pd.DataFrame(data=bc.data, columns=bc.feature_names)
print("Dataype of df_bc: ", type(df_bc))

# Step 3: Add the target column
df_bc['target'] = bc.target

# Step 4: Adding target names
mapped_names = [bc.target_names[t] for t in bc.target]

print("‚úÖ Breast Cancer dataset loaded!")
```

```python
#@title 3.2: Comprehensive EDA

# Displaying dataset shape
print(f"Shape: {df_bc.shape}")

# Displaying first 5 rows
print("\nFirst 5 rows of Wine dataset:")
display(df_bc.head(5))

# Displaying column names
print("\nColoumn Names:")
display(df_bc.columns)

# Displaying data types of all columns
print("\nData types of all coloumns:")
display(df_bc.dtypes)

# Displaying statistical summary
print("\nStatistical Summary:")
display(df_bc.describe())

# Displaying how many samples are in each class
df_bc['target'] = bc.target
print("\nNo. of samples in each class: ")
print(df_bc['target'].value_counts())
```

```python
#@title 3.3: Creating Visualizations

# Creating a bar plot showing the count of each class
print("Creating a bar plot showing the count of each class: ")
df_bc['target'].value_counts().plot(kind='bar')
plt.title('Distribution of Breast Cancer Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Creating a histogram of 'mean radius'
print("\nCreating a histogram of 'mean radius': ")
df_bc['mean radius'].hist()
plt.title('Distribution of Mean Radius')
plt.xlabel('Mean Radius')
plt.ylabel('Frequency')
plt.show()

# Creating a scatter plot of 'mean radius' vs 'mean texture'
print("\nCreating a scatter plot of 'mean radius' vs 'mean texture': ")
plt.scatter(df_bc['mean radius'], df_bc['mean texture'], c=df_bc['target'])
plt.title('Mean Radius vs Mean Texture')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.colorbar(label='Class (0=Malignant, 1=Benign)')
plt.show()
```
